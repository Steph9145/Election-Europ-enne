```python
# Étape 1 : Lire le fichier GeoJSON avec geopandas
gdf = gpd.read_file('circonscriptions-legislatives-p20.geojson')

# Vérifier le CRS actuel
print(f"CRS actuel: {gdf.crs}")
```

```python
# Reprojeter les données en EPSG:3857 (CRS projeté) pour un calcul précis des centroides
gdf_proj = gdf.to_crs(epsg=3857)

# Calculer les coordonnées moyennes des centroides dans le CRS projeté
centroid_y_mean = gdf_proj.geometry.centroid.y.mean()
centroid_x_mean = gdf_proj.geometry.centroid.x.mean()

# Créer un point Shapely avec les coordonnées moyennes
mean_point = Point(centroid_x_mean, centroid_y_mean)

# Créer une GeoSeries avec ce point, puis reprojeter en CRS géographique (EPSG:4326)
mean_point_series = gpd.GeoSeries([mean_point], crs=3857).to_crs(epsg=4326)

# Extraire les coordonnées du point reprojeté
mean_lat = mean_point_series.geometry.y.mean()
mean_lon = mean_point_series.geometry.x.mean()
```

```python
# Étape 2 : Visualiser les données GeoJSON avec folium
# Créer une carte centrée sur les coordonnées moyennes
m = folium.Map(location=[mean_lat, mean_lon], zoom_start=5.5, background_color='red')

# Ajouter les données GeoJSON à la carte
folium.GeoJson(gdf).add_to(m)
# Enregistrer la carte dans un fichier HTML si nécessaire
m.save('map.html')

# Afficher les premières lignes pour vérifier les données
print(gdf.head())
```

```python
# Charger le fichier des résultats par circonscription:
data_result = pd.read_csv('resultats-definitifs-par-circonscription.csv', sep=";")
# Renommer les colonnes pour correspondance au df 'gdf'
data_result.rename(columns={'Code circonscription législative': 'codeCirconscription',
                            'Libellé circonscription législative':'nomCirconscription','Libellé département':'nomDepartement'}, inplace=True)
data_result.info()
```

```python
# On conserve le code circonscription ainsi que le contour 'geometry'
gdf_select = gdf[['codeCirconscription','geometry']]
# Fusion des df
data_merge = pd.merge(data_result,gdf_select,on="codeCirconscription", validate='1:1')
```

```python
# Récupération des lignes 'DOM/TOM" de chaque table
gdf_OM = gdf.tail(20)
data_result_OM=data_result.iloc[539:559]

# Préparation pour fusion - les codes circonscription étant différents d'une table à l'autre
gdf_OM.loc[:,'ForMerge'] = gdf_OM['nomDepartement'] + gdf_OM['nomCirconscription']
data_result_OM.loc[:,'ForMerge'] = data_result_OM['nomDepartement'] + data_result_OM['nomCirconscription']

# On conserve que les contours des circonscription 'geometry'
gdf_select_OM = gdf_OM[['ForMerge','geometry']]
# On fussionne
data_merge_OM = pd.merge(data_result_OM,gdf_select_OM,on="ForMerge", validate='1:1')
# Suppression de la colonne "ForMerge'
data_merge_OM = data_merge_OM.drop(columns=['ForMerge'])

# Concaténation des deux DataFrames
df_combined = pd.concat([data_merge, data_merge_OM])

# Affichage:
df_combined.tail(3)
```

```python
# Vérification des colonnes avant les listes
df_combined.columns[:20]
```

```python
#--------------------------------------------- Mise en valeurs numériques des colonnes 'pourcentage' ---------------------------

# Colonnes récurrentes (par liste)
colonnes1 = [f'% Voix/exprimés {i}' for i in range(1, 39)]
colonnes2 = [f'% Voix/inscrits {i}' for i in range(1, 39)]
colonnes = colonnes1 + colonnes2
# Transformation de chaque colonne
for col in colonnes:
    df_combined[col] = df_combined[col].str.replace(',', '.').str.rstrip('%').astype('float')

# Colonnes fixes
colonnes = [
    '% Votants', '% Abstentions', '% Exprimés/inscrits', '% Exprimés/votants', 
    '% Blancs/inscrits', '% Blancs/votants', '% Nuls/inscrits', '% Nuls/votants'
]
for col in colonnes:
    df_combined[col] = df_combined[col].str.replace(',', '.').str.rstrip('%').astype('float')
```

```python
# Sélection des colonnes utiles pour les listes de candidats et les voix
Selected_columns_nuance = [f'Libellé abrégé de liste {i}' for i in range(1, 39)]
Selected_columns_voix = [f'Voix {i}' for i in range(1, 39)]
Selected_columns_candidats = [f'Libellé de liste {i}' for i in range(1, 39)]
Selected_columns_vote = [f'% Voix/exprimés {i}' for i in range(1, 39)]
Selected_columns_vote_inscrit = [f'% Voix/inscrits {i}' for i in range(1, 39)]

# Extraction des données
List_Candidats = df_combined[Selected_columns_candidats].iloc[0].tolist()
Voix_Candidats = df_combined[Selected_columns_voix].sum().tolist()
nuance_Candidats = df_combined[Selected_columns_nuance].iloc[0].tolist()
Vote_Candidats = df_combined[Selected_columns_vote].mean().tolist()
Vote_inscrits_Candidats = df_combined[Selected_columns_vote_inscrit].mean().tolist()

# Création du DataFrame final
Df = pd.DataFrame({
    'Candidats': List_Candidats,
    'Nuance' : nuance_Candidats,
    'Nb de Voix': Voix_Candidats,
    '% voix_exprimées':Vote_Candidats,
    '% voix_inscrits':Vote_inscrits_Candidats
}, index=range(1, len(List_Candidats) + 1))

# Classement par nombre de voix
Df = Df.sort_values(by='Nb de Voix', ascending=False).round(2)

# Affichage du DataFrame final
Df= Df.reset_index()
Df
```

```python
# Sélectionner les 15 premiers candidats
top_20 = Df.head(20)

# Créer un graphique à barres
plt.figure(figsize=(12, 8))
sns.barplot(x='% voix_exprimées', y='Nuance', data=top_20, palette='viridis')
plt.xlabel('% Voix Exprimées')
plt.ylabel('')
plt.title('Pourcentage de Voix Exprimées par Nuance pour les 15 Candidats ayant reçu le plus de Voix')
plt.show()
```
